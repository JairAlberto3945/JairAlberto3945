MÉTODO DE BOX-MULLER

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def box_muller(n):
    u1 = np.random.rand(n // 2)
    u2 = np.random.rand(n // 2).


    # Fórmulas del método de Box-Muller:
    z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

    # Concatenar los resultados para tener n datos
    return np.concatenate([z1, z2])
```

```python
#Generación y análisis de datos
n = 10000
# Datos simulados
datos = box_muller(n)
# Estadísticas simuladas
media_sim = np.mean(datos)
std_sim = np.std(datos)
# Estadísticas teóricas
media_teo = 0
std_teo = 1

print(f"Media simulada: {media_sim:.4f}")
print(f"Desviación estándar simulada: {std_sim:.4f}")
print(f"Media teórica: {media_teo}")
print(f"Desviación estándar teórica: {std_teo}")
```

```python
x = np.linspace(-5, 5, 1000)
pdf_teorica = stats.norm.pdf(x, loc=media_teo, scale=std_teo)

# Gráfico 1: PDF teórica
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(x, pdf_teorica, color='red', label='PDF teórica')
plt.title("Distribución Normal Estándar (Teórica)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid()

# Gráfico 2: Histograma de datos simulados
plt.subplot(1, 3, 2)
plt.hist(datos, bins=50, density=True, alpha=0.7, color='blue', label='Simulado')
plt.title("Histograma (Datos Simulados)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid()

# Gráfico 3: Comparación teoría vs simulación
plt.subplot(1, 3, 3)
plt.hist(datos, bins=50, density=True, alpha=0.6, label='Simulado')
plt.plot(x, pdf_teorica, color='red', linewidth=2, label='Teórica')
plt.title("Comparación: Teórico vs Simulado")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
