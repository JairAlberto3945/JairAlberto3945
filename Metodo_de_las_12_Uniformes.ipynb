Metodo de las 12 Uniformes

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parámetros de la distribución normal deseada
mu = 5       # media
sigma = 2    # desviación estándar
n = 10000    # número de muestras

# Método de las 12 uniformes
def normal_12_uniformes(n, mu, sigma):
    u = np.random.uniform(0, 1, (n, 12))
    z = np.sum(u, axis=1) - 6  # media 0, varianza 1
    return mu + sigma * z      # escalado a N(mu, sigma^2)
samples = normal_12_uniformes(n, mu, sigma)

# Estadísticas simuladas
media_simulada = np.mean(samples)
std_simulada = np.std(samples)

# Estadísticas teóricas
media_teorica = mu
std_teorica = sigma
```

```python
# Mostrar resultados
print("Media simulada:", media_simulada)
print("Media teórica:", media_teorica)
print("Desviación estándar simulada:", std_simulada)
print("Desviación estándar teórica:", std_teorica)

# Dominio para la curva teórica
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_teorica = norm.pdf(x, loc=mu, scale=sigma)
```

```python
# Gráfica 1: Distribución Teórica
plt.figure(figsize=(8, 4))
plt.plot(x, y_teorica, 'r-', lw=2, label='Teórica')
plt.title("Distribución Normal Teórica")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()

# Gráfica 2: Distribución Simulada
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=50, density=True, alpha=0.7, color='skyblue', label='Simulada')
plt.title("Distribución Normal Simulada (12 Uniformes)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()

# Gráfica 3: Comparación Teórica vs Simulada
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=50, density=True, alpha=0.5, color='skyblue', label='Simulada')
plt.plot(x, y_teorica, 'r-', lw=2, label='Teórica')
plt.title("Comparación: Teórica vs Simulada (12 Uniformes)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
