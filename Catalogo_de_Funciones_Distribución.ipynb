```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from sympy import symbols, simplify, sqrt, latex
from IPython.display import display, Markdown, Math
```

FUNCIONES CONTINUAS


**UNIFORME**

```python
display(Markdown(r"""
## Distribución Uniforme Continua

**Notación estándar**:  \( X \sim \mathcal{U}(a, b) \)

- **a**: límite inferior del intervalo
- **b**: límite superior del intervalo, con \( a < b \)
- **Soporte**: \( x \in [a, b] \)

**Función de densidad (PDF)**:
\[
f(x) = \begin{cases}
\frac{1}{b - a}, & a \leq x \leq b \\
0, & \text{en otro caso}
\end{cases}
\]

**Función de distribución acumulativa (CDF)**:
\[
F(x) = \begin{cases}
0, & x < a \\
\frac{x - a}{b - a}, & a \leq x \leq b \\
1, & x > b
\end{cases}
\]
"""))

#Gráficas de PDF y CDF
a, b = 2, 8
x = np.linspace(a - 2, b + 2, 1000)
pdf_vals = uniform.pdf(x, loc=a, scale=b - a)
cdf_vals = uniform.cdf(x, loc=a, scale=b - a)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Gráfica de PDF
axs[0].plot(x, pdf_vals, color='blue', label='PDF')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# Gráfica de CDF
axs[1].plot(x, cdf_vals, color='green', label='CDF')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
a_sym, b_sym = symbols('a b')
media = simplify((a_sym + b_sym)/2)
varianza = simplify(((b_sym - a_sym)**2)/12)
desv_std = simplify(sqrt(varianza))

media_latex = latex(media)
varianza_latex = latex(varianza)
desv_std_latex = latex(desv_std)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \dfrac{a + b}{2} = " + media_latex))
display(Math(r"\sigma^2 = \dfrac{(b - a)^2}{12} = " + varianza_latex))
display(Math(r"\sigma = \sqrt{\dfrac{(b - a)^2}{12}} " ))


#PSEUDOCODIGO
#Generar u ~ U(0, 1)
#x ← a + (b - a) * u
#Retornar x
```

Código de simulación

```python
def generar_uniforme(a, b, n=1000):
    u = np.random.uniform(0, 1, n)
    return a + (b - a) * u

# Simulación y gráfico de histograma
muestras = generar_uniforme(2, 8, 10000)

plt.hist(muestras, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title("Histograma de muestras simuladas ~ U(2, 8)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

**TRIANGULAR**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang
from sympy import symbols, simplify, sqrt, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Triangular Continua

**Notación estándar**:  \( X \sim \text{Triangular}(a, c, b) \)

- **a**: límite inferior
- **b**: límite superior
- **c**: modo (máximo de la función de densidad), con \( a \leq c \leq b \)
- **Soporte**: \( x \in [a, b] \)

**Función de densidad (PDF)**:
\[
f(x) =
\begin{cases}
0, & x < a \\
\frac{2(x - a)}{(b - a)(c - a)}, & a \leq x < c \\
\frac{2(b - x)}{(b - a)(b - c)}, & c \leq x \leq b \\
0, & x > b
\end{cases}
\]

**Función de distribución acumulativa (CDF)**:
\[
F(x) =
\begin{cases}
0, & x < a \\
\frac{(x - a)^2}{(b - a)(c - a)}, & a \leq x < c \\
1 - \frac{(b - x)^2}{(b - a)(b - c)}, & c \leq x \leq b \\
1, & x > b
\end{cases}
\]
"""))

# Gráficas de PDF y CDF
a, c, b = 2, 5, 8  # parámetros para la triangular
c_param = (c - a) / (b - a)  # modo reescalado para scipy

x = np.linspace(a - 2, b + 2, 1000)
pdf_vals = triang.pdf(x, c_param, loc=a, scale=b - a)
cdf_vals = triang.cdf(x, c_param, loc=a, scale=b - a)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue', label='PDF')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green', label='CDF')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

#Parámetros teóricos
a_sym, c_sym, b_sym = symbols('a c b')
media = simplify((a_sym + b_sym + c_sym)/3)
varianza = simplify((a_sym**2 + b_sym**2 + c_sym**2 - a_sym*b_sym - a_sym*c_sym - b_sym*c_sym)/18)
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \frac{a + b + c}{3} = " + latex(media)))
display(Math(r"\sigma^2 = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{" + latex(varianza) + "}"))

#PSEUDOCODIGO
#Si u < (c - a) / (b - a) entonces:
#x ← a + sqrt( u * (b - a) * (c - a) )
#Si no:
#x ← b - sqrt( (1 - u) * (b - a) * (b - c) )
#Retornar x
```

Código de simulación

```python
def triangular_simulacion(a, c, b, n=10000):
    u = np.random.uniform(0, 1, size=n)
    x = np.where(u < (c - a)/(b - a),
                 a + np.sqrt(u * (b - a) * (c - a)),
                 b - np.sqrt((1 - u) * (b - a) * (b - c)))
    return x

# Simular y graficar histograma comparado con PDF teórica
muestras = triangular_simulacion(a, c, b, n=10000)
x_vals = np.linspace(a, b, 1000)
pdf_teorica = triang.pdf(x_vals, (c - a)/(b - a), loc=a, scale=b - a)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Triangular')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**EXPONENCIAL**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from sympy import symbols, simplify, sqrt, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Exponencial

**Notación estándar**:  \( X \sim \text{Exp}(\lambda) \)

- **\(\lambda > 0\)**: tasa o inverso de la media
- **Soporte**: \( x \in [0, \infty) \)

**Función de densidad (PDF)**:
\[
f(x) = \begin{cases}
\lambda e^{-\lambda x}, & x \geq 0 \\
0, & x < 0
\end{cases}
\]

**Función de distribución acumulativa (CDF)**:
\[
F(x) = \begin{cases}
0, & x < 0 \\
1 - e^{-\lambda x}, & x \geq 0
\end{cases}
\]
"""))

# Gráficas de PDF y CDF
lam = 1.5  # parámetro lambda
x = np.linspace(0, 10, 1000)
pdf_vals = expon.pdf(x, scale=1/lam)
cdf_vals = expon.cdf(x, scale=1/lam)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue', label='PDF')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green', label='CDF')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
λ = symbols('lambda', positive=True)
media = simplify(1 / λ)
varianza = simplify(1 / λ**2)
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \frac{1}{\lambda} = " + latex(media)))
display(Math(r"\sigma^2 = \frac{1}{\lambda^2} = " + latex(varianza)))
display(Math(r"\sigma = \frac{1}{\lambda} = " + latex(desv_std)))

#PSEUDOCODIGO
#Generar u ~ U(0, 1)
#x ← -ln(1 - u) / λ
#Retornar x
```

Código de simulación

```python
def exponencial_simulacion(lam, n=10000):
    u = np.random.uniform(0, 1, size=n)
    x = -np.log(1 - u) / lam
    return x

# Simulación
muestras = exponencial_simulacion(lam, n=10000)
x_vals = np.linspace(0, 10, 1000)
pdf_teorica = expon.pdf(x_vals, scale=1/lam)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Exponencial')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**M-Erlang**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import erlang
from sympy import symbols, simplify, sqrt, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Erlang (m-Erlang)

**Notación estándar**:
\( X \sim \text{Erlang}(m, \lambda) \)

- \( m \in \mathbb{N} \): número de etapas o fases
- \( \lambda > 0 \): tasa por etapa
- **Soporte**: \( x \in [0, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m - 1)!}, \quad x \geq 0
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = 1 - \sum_{k=0}^{m-1} \frac{(\lambda x)^k e^{-\lambda x}}{k!}, \quad x \geq 0
$$
"""))

# Gráficas de PDF y CDF
m = 3
lam = 1.2
x = np.linspace(0, 15, 1000)

pdf_vals = erlang.pdf(x, a=m, scale=1/lam)
cdf_vals = erlang.cdf(x, a=m, scale=1/lam)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
m_sym, λ = symbols('m lambda', integer=True, positive=True)
media = simplify(m_sym / λ)
varianza = simplify(m_sym / λ**2)
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \frac{m}{\lambda} = " + latex(media)))
display(Math(r"\sigma^2 = \frac{m}{\lambda^2} = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{\frac{m}{\lambda^2}} = " + latex(desv_std)))

#PESUDOCODIGO
#Inicializar x ← 0
#Repetir m veces:
#Generar u ~ U(0, 1)
#x ← x - ln(u) / λ
#Retornar x
```

Código de Simulación

```python
def erlang_simulacion(m, lam, n=10000):
    u = np.random.uniform(0, 1, size=(n, m))
    x = -np.log(u).sum(axis=1) / lam
    return x

# Simulación
muestras = erlang_simulacion(m, lam)
x_vals = np.linspace(0, 15, 1000)
pdf_teorica = erlang.pdf(x_vals, a=m, scale=1/lam)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución m-Erlang')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**GAMMA**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from sympy import symbols, simplify, sqrt, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Gamma

**Notación estándar**:
\( X \sim \text{Gamma}(\alpha, \lambda) \)

- \( \alpha > 0 \): parámetro de forma
- \( \lambda > 0 \): tasa (inversa de escala)
- **Soporte**: \( x \in [0, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{\lambda^\alpha x^{\alpha - 1} e^{-\lambda x}}{\Gamma(\alpha)}, \quad x \geq 0
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = \int_0^x f(t)\,dt = \frac{\gamma(\alpha, \lambda x)}{\Gamma(\alpha)}, \quad x \geq 0
$$
donde \( \gamma(\alpha, \lambda x) \) es la función gamma incompleta inferior.
"""))

# Gráficas de PDF y CDF
alpha = 3.5
lam = 1.0
x = np.linspace(0, 20, 1000)

pdf_vals = gamma.pdf(x, a=alpha, scale=1/lam)
cdf_vals = gamma.cdf(x, a=alpha, scale=1/lam)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
α, λ = symbols('alpha lambda', positive=True)
media = simplify(α / λ)
varianza = simplify(α / λ**2)
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \frac{\alpha}{\lambda} = " + latex(media)))
display(Math(r"\sigma^2 = \frac{\alpha}{\lambda^2} = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{\frac{\alpha}{\lambda^2}} = " + latex(desv_std)))

#PSEUDOCODIGO
#Si α es entero (α = m):
#Inicializar x ← 0
#Repetir m veces:
#Generar u ~ U(0, 1)
#x ← x - ln(u) / λ
#Retornar x
#Si α no es entero:
#Usar el algoritmo de Ahrens-Dieter o Marsaglia-Tsang
```

Código de simulación

```python
def gamma_simulacion_entera(alpha, lam, n=10000):
    m = int(alpha)
    u = np.random.uniform(0, 1, size=(n, m))
    x = -np.log(u).sum(axis=1) / lam
    return x

# Simulación (solo si alpha es entero)
alpha_ent = 3  # usar valor entero para simulación básica
muestras = gamma_simulacion_entera(alpha_ent, lam)
x_vals = np.linspace(0, 20, 1000)
pdf_teorica = gamma.pdf(x_vals, a=alpha_ent, scale=1/lam)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Gamma (α entero)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**WEIBULL**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
from sympy import symbols, simplify, sqrt, gamma as gamma_func, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Weibull

**Notación estándar**:
\( X \sim \text{Weibull}(k, \lambda) \)

- \( k > 0 \): parámetro de forma
- \( \lambda > 0 \): parámetro de escala
- **Soporte**: \( x \in [0, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{k}{\lambda} \left( \frac{x}{\lambda} \right)^{k - 1} e^{-(x/\lambda)^k}, \quad x \geq 0
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = 1 - e^{-(x/\lambda)^k}, \quad x \geq 0
$$
"""))

# Gráficas de PDF y CDF
k = 1.5  # forma
lam = 2  # escala
x = np.linspace(0, 10, 1000)

pdf_vals = weibull_min.pdf(x, c=k, scale=lam)
cdf_vals = weibull_min.cdf(x, c=k, scale=lam)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
k_sym, λ = symbols('k lambda', positive=True)
media = simplify(λ * gamma_func(1 + 1 / k_sym))
varianza = simplify(λ**2 * (gamma_func(1 + 2 / k_sym) - gamma_func(1 + 1 / k_sym)**2))
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \lambda \, \Gamma\left(1 + \frac{1}{k} \right) = " + latex(media)))
display(Math(r"\sigma^2 = \lambda^2 \left[\Gamma\left(1 + \frac{2}{k} \right) - \Gamma^2\left(1 + \frac{1}{k} \right) \right] = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{\sigma^2} = " + latex(desv_std)))
#PSEUDOCODIGO
#Generar u ~ U(0, 1)
#x ← λ * (-ln(1 - u))^(1/k)
#Retornar x
```

Código de Simulación

```python
def weibull_simulacion(k, lam, n=10000):
    u = np.random.uniform(0, 1, n)
    x = lam * (-np.log(1 - u))**(1 / k)
    return x

# Simulación
muestras = weibull_simulacion(k, lam)
x_vals = np.linspace(0, 10, 1000)
pdf_teorica = weibull_min.pdf(x_vals, c=k, scale=lam)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Weibull')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**NORMAL**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sympy import symbols, simplify, sqrt, pi, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Normal

**Notación estándar**:
\( X \sim \mathcal{N}(\mu, \sigma^2) \)

- \( \mu \in \mathbb{R} \): media
- \( \sigma > 0 \): desviación estándar
- **Soporte**: \( x \in (-\infty, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{x - \mu}{\sigma \sqrt{2}} \right) \right]
$$
donde \( \text{erf} \) es la función de error.
"""))

# Gráficas de PDF y CDF
mu = 0
sigma = 1
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

pdf_vals = norm.pdf(x, loc=mu, scale=sigma)
cdf_vals = norm.cdf(x, loc=mu, scale=sigma)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
mu_sym, sigma_sym = symbols('mu sigma', positive=True)
media = simplify(mu_sym)
varianza = simplify(sigma_sym**2)
desv_std = simplify(sigma_sym)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = " + latex(media)))
display(Math(r"\sigma^2 = " + latex(varianza)))
display(Math(r"\sigma = " + latex(desv_std)))
#PSEUDOCODIGO
#Generar u1, u2 ~ U(0, 1)
#z ← sqrt(-2 * ln(u1)) * cos(2π * u2)
#x ← μ + σ * z
#Retornar x
```

Código de simulación

```python
def normal_simulacion(mu, sigma, n=10000):
    u1 = np.random.uniform(0, 1, n)
    u2 = np.random.uniform(0, 1, n)
    z = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    x = mu + sigma * z
    return x

# Simulación
muestras = normal_simulacion(mu, sigma)
x_vals = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf_teorica = norm.pdf(x_vals, loc=mu, scale=sigma)

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**LOG-NORMAL**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm
from sympy import symbols, simplify, sqrt, exp, pi, latex
from IPython.display import display, Markdown, Math
display(Markdown(r"""
## Distribución Log-normal

**Notación estándar**:
\( X \sim \text{Lognormal}(\mu, \sigma^2) \)

- \( \mu \in \mathbb{R} \): media del logaritmo
- \( \sigma > 0 \): desviación estándar del logaritmo
- **Soporte**: \( x \in (0, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{1}{x \, \sigma \sqrt{2\pi}} \, \exp\left( -\frac{(\ln x - \mu)^2}{2\sigma^2} \right), \quad x > 0
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = \Phi \left( \frac{\ln x - \mu}{\sigma} \right), \quad x > 0
$$
donde \( \Phi \) es la CDF de la distribución normal estándar.
"""))

# Gráficas de PDF y CDF
mu, sigma = 0, 0.5
x = np.linspace(0.01, 5, 1000)

pdf_vals = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
cdf_vals = lognorm.cdf(x, s=sigma, scale=np.exp(mu))

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
mu_sym, sigma_sym = symbols('mu sigma', positive=True)
media = simplify(exp(mu_sym + sigma_sym**2 / 2))
varianza = simplify((exp(sigma_sym**2) - 1) * exp(2 * mu_sym + sigma_sym**2))
desv_std = simplify(sqrt(varianza))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \mathbb{E}[X] = " + latex(media)))
display(Math(r"\sigma^2 = \text{Var}(X) = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{\text{Var}(X)} = " + latex(desv_std)))

#PSEUDOCODIGO
#Generar z ~ N(μ, σ²)
#x ← exp(z)
#Retornar x
```

Código de simulación

```python
def lognormal_simulacion(mu, sigma, n=10000):
    z = np.random.normal(loc=mu, scale=sigma, size=n)
    x = np.exp(z)
    return x

# Simulación
muestras = lognormal_simulacion(mu, sigma)
x_vals = np.linspace(0.01, 5, 1000)
pdf_teorica = lognorm.pdf(x_vals, s=sigma, scale=np.exp(mu))

plt.figure(figsize=(8, 4))
plt.hist(muestras, bins=50, density=True, alpha=0.6, label='Simulación')
plt.plot(x_vals, pdf_teorica, color='red', lw=2, label='PDF teórica')
plt.title('Simulación vs PDF Teórica - Distribución Log-normal')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

**BETA**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from sympy import symbols, simplify, gamma, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Beta

**Notación estándar**:
\( X \sim \text{Beta}(\alpha, \beta) \)

- \( \alpha > 0 \): parámetro de forma
- \( \beta > 0 \): parámetro de forma
- **Soporte**: \( x \in (0, 1) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha)\Gamma(\beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}, \quad 0 < x < 1
$$

**Función de distribución acumulativa (CDF)**:
No tiene forma cerrada general. Se calcula numéricamente con funciones especiales incompletas de beta.
"""))

# Gráficas de PDF y CDF
a, b = 2.0, 5.0  # parámetros alpha y beta
x = np.linspace(0.001, 0.999, 1000)

pdf_vals = beta.pdf(x, a, b)
cdf_vals = beta.cdf(x, a, b)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x, pdf_vals, color='blue')
axs[0].fill_between(x, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# Parámetros teóricos
alpha, beta_sym = symbols('alpha beta', positive=True)
media = simplify(alpha / (alpha + beta_sym))
varianza = simplify((alpha * beta_sym) / ((alpha + beta_sym)**2 * (alpha + beta_sym + 1)))

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mu = \mathbb{E}[X] = " + latex(media)))
display(Math(r"\sigma^2 = \text{Var}(X) = " + latex(varianza)))
#PSEUDOCÓDIGO
#Generar X₁ ~ Gamma(α, 1)
#Generar X₂ ~ Gamma(β, 1)
#x ← X₁ / (X₁ + X₂)
#Retornar x
```

Código de simulación

```python
def simular_beta(alpha, beta_param, size=10000):
    x1 = np.random.gamma(shape=alpha, scale=1.0, size=size)
    x2 = np.random.gamma(shape=beta_param, scale=1.0, size=size)
    return x1 / (x1 + x2)

# Simular y graficar
samples = simular_beta(a, b)

plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title('Histograma de muestras simuladas (Beta)')
plt.xlabel('x')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.show()
```

**LOG-LOGISTICA**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import fisk  # Distribución log-logística
from sympy import symbols, simplify, sqrt, latex, pi, sin
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Log-Logística

**Notación estándar**:
\( X \sim \text{LogLogística}(\alpha, \beta) \)

- \( \alpha > 0 \): parámetro de forma
- \( \beta > 0 \): parámetro de escala
- **Soporte**: \( x \in (0, \infty) \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{(\alpha/\beta)(x/\beta)^{\alpha - 1}}{\left[1 + (x/\beta)^{\alpha}\right]^2}, \quad x > 0
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = \frac{1}{1 + (x/\beta)^{-\alpha}}, \quad x > 0
$$

**Nota**:
- La media existe solo si \( \alpha > 1 \)
- La varianza existe solo si \( \alpha > 2 \)
"""))

# Parámetros teóricos simbólicos
alpha, beta = symbols('alpha beta', positive=True)

# Media
media = simplify(beta * pi / (alpha * sin(pi / alpha)))

# Varianza (cuando existe)
varianza = simplify(
    beta**2 * ((2 * pi / (alpha * sin(2 * pi / alpha))) - (pi / (alpha * sin(pi / alpha)))**2)
)

# Desviación estándar
desv_std = simplify(sqrt(varianza))

# Mostrar resultados
display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = \beta \cdot \frac{\pi}{\alpha \sin(\pi/\alpha)} = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = \beta^2 \left[ \frac{2\pi}{\alpha \sin(2\pi/\alpha)} - \left( \frac{\pi}{\alpha \sin(\pi/\alpha)} \right)^2 \right] = " + latex(varianza)))
display(Math(r"\sigma = \sqrt{\mathrm{Var}(X)} = " + latex(desv_std)))

#Gráficas PDF y CDF
alpha_val = 3  # α > 2 para que existan media y varianza
beta_val = 1

x_vals = np.linspace(0.01, 10, 1000)

# Funciones de densidad y distribución acumulativa
pdf_vals = fisk.pdf(x_vals, c=alpha_val, scale=beta_val)
cdf_vals = fisk.cdf(x_vals, c=alpha_val, scale=beta_val)

# Gráficos
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x_vals, pdf_vals, color='blue')
axs[0].fill_between(x_vals, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x_vals, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# PSEUDOCÓDIGO
# Generar U ~ Uniforme(0, 1)
# x ← β * (U / (1 - U))^(1/α)
# Retornar x
```

Codigo de simulación

```python
def simular_loglogistica(alpha, beta, n=10000):
    """
    Simula n valores de una distribución log-logística
    mediante la técnica de la función inversa.
    """
    u = np.random.uniform(0, 1, size=n)
    x = beta * (u / (1 - u))**(1 / alpha)
    return x

alpha_val = 3  # Forma
beta_val = 1   # Escala
muestras = simular_loglogistica(alpha_val, beta_val, n=10000)
# --- Comparación visual con la PDF teórica ---
x_vals = np.linspace(0.01, 10, 1000)
pdf_teorica = fisk.pdf(x_vals, c=alpha_val, scale=beta_val)

# --- Gráfico ---
plt.figure(figsize=(10, 5))
plt.hist(muestras, bins=100, density=True, alpha=0.5, label='Simulación', color='skyblue')
plt.plot(x_vals, pdf_teorica, 'r-', lw=2, label='PDF teórica')
plt.title('Simulación vs. PDF Teórica - Distribución Log-Logística')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
```

**JOHNSON**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import johnsonsu
from sympy import symbols, simplify, sqrt, sinh, exp, pi, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Johnson SU

**Notación estándar**:
\( X \sim \text{JohnsonSU}(\gamma, \delta, \xi, \lambda) \)

- \( \gamma \in \mathbb{R} \): parámetro de ubicación
- \( \delta > 0 \): parámetro de forma
- \( \xi \in \mathbb{R} \): parámetro de traslación (shift)
- \( \lambda > 0 \): parámetro de escala
- **Soporte**: \( x \in \mathbb{R} \)

**Función de densidad (PDF)**:
$$
f(x) = \frac{\delta}{\lambda \sqrt{2\pi} \sqrt{1 + z^2}} \, \exp\left( -\frac{1}{2} (\gamma + \delta \, \sinh^{-1}(z))^2 \right), \quad z = \frac{x - \xi}{\lambda}
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = \Phi \left( \gamma + \delta \, \sinh^{-1}\left( \frac{x - \xi}{\lambda} \right) \right)
$$
donde \( \Phi \) es la CDF de la normal estándar.
"""))


# Parámetros teóricos simbólicos

gamma_sym, delta_sym, xi_sym, lambda_sym = symbols('gamma delta xi lambda', real=True, positive=True)

# Media y varianza (expresadas en función de los parámetros)
media = simplify(xi_sym + lambda_sym * sinh(gamma_sym / delta_sym))
# Varianza: expresión teórica es compleja, por eso se omite aquí

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Markdown("La varianza teórica tiene una expresión más compleja y depende del momento de segundo orden de la transformación."))

# Definir parámetros numéricos
gamma_val = 1.0
delta_val = 2.0
xi_val = 0.0
lambda_val = 1.0

# Generar valores x y calcular PDF/CDF
x_vals = np.linspace(-5, 5, 1000)
pdf_vals = johnsonsu.pdf(x_vals, a=gamma_val, b=delta_val, loc=xi_val, scale=lambda_val)
cdf_vals = johnsonsu.cdf(x_vals, a=gamma_val, b=delta_val, loc=xi_val, scale=lambda_val)

# Graficar
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PDF
axs[0].plot(x_vals, pdf_vals, color='blue')
axs[0].fill_between(x_vals, pdf_vals, alpha=0.2)
axs[0].set_title('Función de Densidad (PDF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(True)

# CDF
axs[1].plot(x_vals, cdf_vals, color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()


#PSEUDOCODIGO
#Generar Z ~ Normal(0, 1)
#x ← ξ + λ * sinh((Z - γ) / δ)
#Retornar x
```

Código de Simulación

```python
def simular_johnsonsu(gamma, delta, xi, lambd, n=10000):
    """
    Simula n valores de una distribución Johnson SU
    mediante la transformación inversa desde una normal estándar.
    """
    z = np.random.normal(0, 1, size=n)
    x = xi + lambd * np.sinh((z - gamma) / delta)
    return x

# Simulación
sim_data = simular_johnsonsu(gamma_val, delta_val, xi_val, lambda_val, n=10000)

# Comparación visual
plt.figure(figsize=(10, 5))
plt.hist(sim_data, bins=100, density=True, alpha=0.5, label='Simulación', color='skyblue')
plt.plot(x_vals, pdf_vals, 'r-', label='PDF teórica')
plt.title('Simulación vs. PDF Teórica - Johnson SU')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint
from sympy import symbols, simplify, Rational, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Uniforme Discreta

**Notación estándar**:
\( X \sim \text{Uniforme}(a, b) \), donde \( a, b \in \mathbb{Z}, a < b \)

- \( a \): menor valor entero
- \( b \): mayor valor entero
- Soporte: \( x \in \{a, a+1, \dots, b\} \)

**Función de masa de probabilidad (PMF)**:
$$
P(X = x) = \frac{1}{b - a + 1}, \quad x = a, a+1, \dots, b
$$

**Función de distribución acumulativa (CDF)**:
$$
F(x) = P(X \leq x) = \frac{\lfloor x \rfloor - a + 1}{b - a + 1}, \quad \text{para } x \in [a, b]
$$
"""))

# Parámetros teóricos simbólicos

a, b = symbols('a b', integer=True)
n = b - a + 1

media = simplify((a + b) / 2)
varianza = simplify((n**2 - 1) / 12)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = " + latex(varianza)))


# Gráficas PMF y CDF

a_val = 2
b_val = 10

x_vals = np.arange(a_val, b_val + 1)
pmf_vals = randint.pmf(x_vals, low=a_val, high=b_val + 1)
cdf_vals = randint.cdf(x_vals, low=a_val, high=b_val + 1)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PMF
axs[0].stem(x_vals, pmf_vals, basefmt=" ")
axs[0].set_title('Función de Masa de Probabilidad (PMF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(X = x)')
axs[0].grid(True)

# CDF
axs[1].step(x_vals, cdf_vals, where='post', color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()
#PSEUDOCODIGO
#ef simular_uniforme_discreta(a, b, n=10000):
#u = np.random.uniform(0, 1, n)
#x = a + np.floor((b - a + 1) * u).astype(int)
#return x
```

**BERNOULLI**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
from sympy import symbols, simplify, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Bernoulli

**Notación estándar**
\( X \sim \text{Bernoulli}(p) \)

- \( p \in (0,1) \): probabilidad de éxito
- Soporte: \( x \in \{0, 1\} \)

**Función de masa de probabilidad (PMF)**
$$
f(x) = P(X = x) =
  \begin{cases}
    1-p, & x = 0,\\[4pt]
    p, & x = 1.
  \end{cases}
$$

**Función de distribución acumulativa (CDF)**
$$
F(x) = P(X \le x) =
  \begin{cases}
    0, & x < 0,\\[4pt]
    1-p, & 0 \le x < 1,\\[4pt]
    1, & x \ge 1.
  \end{cases}
$$
"""))


# Parámetros teóricos simbólicos

p = symbols('p', positive=True, real=True)

media = simplify(p)              # E[X] = p
varianza = simplify(p * (1 - p)) # Var[X] = p(1-p)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = " + latex(varianza)))

# Gráficas PMF y CDF

p_val = 0.3            # Probabilidad de éxito
x_vals = np.array([0, 1])

pmf_vals = bernoulli.pmf(x_vals, p_val)
cdf_vals = bernoulli.cdf(x_vals, p_val)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Densidad (PMF)
axs[0].bar(x_vals, pmf_vals, width=0.4, color='skyblue', edgecolor='black')
axs[0].set_xticks(x_vals)
axs[0].set_title('Función de Densidad (PMF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].grid(axis='y')

# CDF
axs[1].step(x_vals, cdf_vals, where='post', color='green')
axs[1].set_xticks(x_vals)
axs[1].set_ylim(-0.05, 1.05)
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# PSEUDOCÓDIGO
# Entrada: p (probabilidad de éxito)
# Generar u ~ Uniforme(0, 1)
# Si u < p entonces x ← 1
# En caso contrario x ← 0
# Retornar x
```

Çódigo de simulación

```python
def simular_bernoulli(p, n=10000):
    """
    Simula n observaciones de una variable aleatoria Bernoulli(p)

    Parámetros:
    - p: probabilidad de éxito (0 < p < 1)
    - n: número de muestras a generar (default = 10,000)

    Retorna:
    - arreglo con 0s y 1s representando los resultados
    """
    u = np.random.uniform(0, 1, n)  # Generar números aleatorios uniformes
    x = (u < p).astype(int)         # Éxito si u < p → 1, fallo → 0
    return x

# Parámetro y número de simulaciones
p_val = 0.3
n_muestras = 10000

# Simulación
muestras = simular_bernoulli(p_val, n=n_muestras)

# Frecuencia relativa
valores, frecs = np.unique(muestras, return_counts=True)
probs_empiricas = frecs / frecs.sum()

# Gráfico de barras
plt.figure(figsize=(6, 4))
plt.bar(valores, probs_empiricas, width=0.4, color='orange', edgecolor='black')
plt.xticks([0, 1])
plt.title(f'Simulación Bernoulli(p={p_val}) con {n_muestras} muestras')
plt.xlabel('x')
plt.ylabel('Frecuencia relativa')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

**GEOMETRICA**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom
from sympy import symbols, simplify, latex
from IPython.display import display, Markdown, Math


display(Markdown(r"""
## Distribución Geométrica

**Notación estándar**
\( X \sim \text{Geom}(p) \)

- \( p \in (0,1) \): probabilidad de éxito
- Soporte: \( x \in \{1, 2, 3, \dots\} \)

**Función de masa de probabilidad (PMF)**
$$
P(X = x) = (1-p)^{x-1} p, \quad x = 1, 2, 3, \dots
$$

**Función de distribución acumulativa (CDF)**
$$
F(x) = P(X \le x) = 1 - (1 - p)^x, \quad x \in \mathbb{N}
$$
"""))


# Parámetros teóricos simbólicos

p = symbols('p', positive=True, real=True)

media = simplify(1 / p)
varianza = simplify((1 - p) / p**2)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = " + latex(varianza)))

# Gráficas PMF y CDF

p_val = 0.3
x_vals = np.arange(1, 20)

pmf_vals = geom.pmf(x_vals, p=p_val)
cdf_vals = geom.cdf(x_vals, p=p_val)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PMF
axs[0].stem(x_vals, pmf_vals, basefmt=" ")  # <--- corregido aquí
axs[0].set_title('Función de Masa de Probabilidad (PMF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(X = x)')
axs[0].grid(True)

# CDF
axs[1].step(x_vals, cdf_vals, where='post', color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# PSEUDOCÓDIGO
# Entrada: p (probabilidad de éxito)
# Generar u ~ Uniforme(0, 1)
# x ← ⌈ ln(1 - u) / ln(1 - p) ⌉
# Retornar x
```

Código de Simulación

```python
def simular_geometrica(p, n=10000):
    """
    Simula n observaciones de una variable aleatoria Geométrica(p)
    (número de ensayos hasta el primer éxito, incluyendo el éxito)
    """
    u = np.random.uniform(0, 1, n)
    x = np.ceil(np.log(1 - u) / np.log(1 - p)).astype(int)
    return x

# Parámetros de simulación
p_val = 0.3           # Probabilidad de éxito
n_muestras = 10000    # Número de simulaciones

# Ejecutar simulación
muestras = simular_geometrica(p_val, n_muestras)

# Calcular frecuencia relativa
valores, frecs = np.unique(muestras, return_counts=True)
probs_empiricas = frecs / frecs.sum()

# Gráfico de barras de la simulación
plt.figure(figsize=(8, 4))
plt.bar(valores, probs_empiricas, width=0.8, color='skyblue', edgecolor='black')
plt.title(f'Simulación de la Geométrica (p = {p_val})')
plt.xlabel('Número de ensayos hasta el primer éxito')
plt.ylabel('Frecuencia relativa')
plt.grid(axis='y')
plt.xlim(0.5, min(max(valores) + 0.5, 25))  # Límite visual en eje x
plt.show()
```

**BINOMIAL NEGATIVA**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom
from sympy import symbols, simplify, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución Binomial Negativa

**Notación estándar**
\( X \sim \text{BinomialNegativa}(r, p) \)

- \( r \in \mathbb{N} \): número de éxitos deseados
- \( p \in (0,1) \): probabilidad de éxito en cada ensayo
- Soporte: \( x \in \{0, 1, 2, \dots\} \) (número de fracasos antes del r-ésimo éxito)

**Función de masa de probabilidad (PMF)**
$$
P(X = x) = \binom{x + r - 1}{x} p^r (1 - p)^x
$$

**Función de distribución acumulativa (CDF)**
No tiene forma cerrada sencilla, pero se calcula numéricamente.
"""))

# Parámetros teóricos simbólicos

r, p = symbols('r p', positive=True, integer=True)
media = simplify(r * (1 - p) / p)
varianza = simplify(r * (1 - p) / p**2)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = " + latex(varianza)))

# Gráficas PMF y CDF

r_val = 3
p_val = 0.4
x_vals = np.arange(0, 20)

pmf_vals = nbinom.pmf(x_vals, n=r_val, p=p_val)
cdf_vals = nbinom.cdf(x_vals, n=r_val, p=p_val)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Densidad
axs[0].bar(x_vals, pmf_vals, color='skyblue', edgecolor='black')
axs[0].set_title('Función de Densidad (PMF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(X = x)')
axs[0].grid(axis='y')

# CDF
axs[1].step(x_vals, cdf_vals, where='post', color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# PSEUDOCÓDIGO
# Entrada: r (número de éxitos), p (probabilidad de éxito)
# x ← 0
# repeticiones ← 0
# mientras repeticiones < r hacer
#     Generar u ~ Uniforme(0, 1)
#     Si u < p entonces
#         repeticiones ← repeticiones + 1
#     sino
#         x ← x + 1
# Retornar x (número de fracasos antes de r éxitos)
```

Código de simulacion

```python
def simular_binomial_negativa(r, p, n_sim=10000):
    """
    Simula una variable aleatoria con distribución Binomial Negativa.

    Parámetros:
    - r: número de éxitos deseados
    - p: probabilidad de éxito en un ensayo
    - n_sim: número de simulaciones

    Retorna:
    - Un arreglo con el número de fracasos antes de obtener r éxitos
    """
    resultados = []
    for _ in range(n_sim):
        fracasos = 0
        exitos = 0
        while exitos < r:
            u = np.random.uniform()
            if u < p:
                exitos += 1
            else:
                fracasos += 1
        resultados.append(fracasos)
    return np.array(resultados)

r_val = 3     # número de éxitos
p_val = 0.4   # probabilidad de éxito

sim_data = simular_binomial_negativa(r_val, p_val)

# Graficar histograma de resultados simulados
plt.hist(sim_data, bins=np.arange(0, max(sim_data)+2) - 0.5, density=True, alpha=0.7,
         color='orange', edgecolor='black')
plt.title("Simulación de la Binomial Negativa")
plt.xlabel("Fracasos hasta obtener r éxitos")
plt.ylabel("Frecuencia relativa")
plt.grid(True)
plt.show()
```

**POISSON**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from sympy import symbols, simplify, latex
from IPython.display import display, Markdown, Math

display(Markdown(r"""
## Distribución de Poisson

**Notación estándar**
\( X \sim \text{Poisson}(\lambda) \)

- \( \lambda > 0 \): tasa de ocurrencia (media esperada)
- Soporte: \( x \in \mathbb{N}_0 = \{0, 1, 2, \dots\} \)

**Función de masa de probabilidad (PMF)**
$$
P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}, \quad x = 0, 1, 2, \dots
$$

**Función de distribución acumulativa (CDF)**
$$
F(x) = P(X \le x) = \sum_{k=0}^{\lfloor x \rfloor} \frac{e^{-\lambda} \lambda^k}{k!}
$$
"""))


#Parámetros teóricos simbólicos

λ = symbols('lambda', positive=True, real=True)

media = simplify(λ)
varianza = simplify(λ)

display(Markdown("### Parámetros Teóricos"))
display(Math(r"\mathbb{E}[X] = " + latex(media)))
display(Math(r"\mathrm{Var}(X) = " + latex(varianza)))

# Gráficas PMF y CDF

lambda_val = 4  # Valor de lambda
x_vals = np.arange(0, 16)

pmf_vals = poisson.pmf(x_vals, mu=lambda_val)
cdf_vals = poisson.cdf(x_vals, mu=lambda_val)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# PMF
axs[0].stem(x_vals, pmf_vals, basefmt=" ")
axs[0].set_title('Función de Masa de Probabilidad (PMF)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('P(X = x)')
axs[0].grid(True)

# CDF
axs[1].step(x_vals, cdf_vals, where='post', color='green')
axs[1].set_title('Función de Distribución Acumulativa (CDF)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('F(x)')
axs[1].grid(True)

plt.tight_layout()
plt.show()

# PSEUDOCÓDIGO
# Entrada: λ (tasa de ocurrencia)
# Generar L ← exp(–λ)
# Inicializar:
#     k ← 0
#     p ← 1
# Repetir:
#     Generar u ~ Uniforme(0, 1)
#     p ← p * u
#     k ← k + 1
# Mientras p > L
# Retornar x ← k - 1
```

Código de simulación

```python
def simular_poisson(lambd, n=10000):
    muestras = []
    for _ in range(n):
        L = np.exp(-lambd)
        k = 0
        p = 1
        while p > L:
            u = np.random.uniform(0, 1)
            p *= u
            k += 1
        muestras.append(k - 1)
    return np.array(muestras)

# ------------------------------
# Simulación y comparación con PMF teórica
# ------------------------------
lambda_val = 4
np.random.seed(0)
simulaciones = simular_poisson(lambda_val, n=10000)

# Valores teóricos
x_vals = np.arange(0, 16)
pmf_teorica = poisson.pmf(x_vals, mu=lambda_val)

# Histograma
plt.figure(figsize=(8, 4))
plt.hist(simulaciones, bins=np.arange(0, 17)-0.5, density=True,
         alpha=0.7, color='orange', edgecolor='black', label='Simulación')
plt.plot(x_vals, pmf_teorica, 'o-', color='blue', label='PMF Teórica')
plt.title('Simulación de la Distribución de Poisson')
plt.xlabel('x')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
